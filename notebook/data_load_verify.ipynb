```python
import pandas as pd
import boto3
from io import StringIO

s3 = boto3.client('s3')
bucket_name = 'cmse492-cranda98-nyc311-964165949460-us-east-1-an'
file_name = 'modeling/modeling_data.csv'

obj = s3.get_object(Bucket=bucket_name, Key=file_name)
data = obj['Body'].read().decode('utf-8')
df = pd.read_csv(StringIO(data))
print('worked')
```

```python
# First few rows
print("First few rows:")
print(df.head())

# Data types and missing values
print("\n" + "="*50)
print("Data types and missing values:")
print(df.info())
```

```python
# Summary statistics
print("Summary statistics:")
print(df.describe())
```

```python
# Check target balance
print("Target distribution (complaint_category):")
print(df['complaint_category'].value_counts().sort_index())

# As percentage
print("\nTarget distribution (%)")
print((df['complaint_category'].value_counts() / len(df) * 100).sort_index().round(2))
```
